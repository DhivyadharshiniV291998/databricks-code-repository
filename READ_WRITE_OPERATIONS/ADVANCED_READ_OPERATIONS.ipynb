##there are 3 typical read modes (default- Permissive)<br>
1. permissive — All fields are set to null and corrupted records are placed in a string column called _corrupt_record<br>
2. dropMalformed — Drops all rows containing corrupt records.<br>
3. failFast — Fails when corrupt records are encountered.


#csv file format:

##VERY IMPORTANT:
path, schema, sep, header, inferSchema, samplingRatio, mode, columnNameOfCorruptRecord

```python
#mode used - dropMalformed- when encountered with malformed data, drop the malformed data(did not go with remaining records of the table)
struct1="cid int,name string,amt decimal(10,2),dop date"
df1=spark.read.schema(struct1).csv("/Volumes/workspace/default/volumewe47_datalake/malformeddata.txt",header=False,sep=",",comment="#",mode="dropMalformed")
display(df1)
```

```python
#mode used - permissive- when encountered with malformed data, keep the malformed data in a column called corrupt_record
struct1="cid int,name string,amt decimal(10,2),dop date,corrupt_record string"
df1=spark.read.schema(struct1).csv("/Volumes/workspace/default/volumewe47_datalake/malformeddata.txt",header=False,sep=",",comment='#',mode='permissive',columnNameOfCorruptRecord='corrupt_record')
display(df1)
display(df1.where("corrupt_record is not null"))
```

##IMPORTANT:
quote, escape,comment, ignoreLeadingWhiteSpace, ignoreTrailingWhiteSpace, nullValue, nanValue, dateFormat, timestampFormat, multiLine, enforceSchema,pathGlobFilter, recursiveFileLookup,modifiedBefore,modifiedAfter

```python
struct1="cid int,name string,amt decimal(10,2),dop date"
df1=spark.read.schema(struct1).csv("/Volumes/workspace/default/volumewe47_datalake/malformeddata1.txt",header=False,sep=",",comment='#',quote="'",escape="|",ignoreLeadingWhiteSpace=True,ignoreTrailingWhiteSpace=True,dateFormat="yyyy-dd-MM",mode="permissive",modifiedAfter="2025-12-26",multiLine=True,nullValue="na")
display(df1)
```

#json file format:

##VERY IMPORTANT:

path,schema,columnNameOfCorruptRecord,dateFormat,timestampFormat,multiLine,pathGlobFilter,recursiveFileLookup<BR>
**No header, No inferSchema, No sep in json**...

```python
strt1="id int,name string,amt float, dop date,corrupted_record string"
df_json_multiline=spark.read.json("/Volumes/workspace/default/volumewe47_datalake/simple_json_multiline3.txt",allowComments=True,allowSingleQuotes=True,allowUnquotedFieldNames=True,columnNameOfCorruptRecord="corrupted_record",dateFormat="yyyy-dd-MM",multiLine=True,schema=strt1)
display(df_json_multiline)
df_json_multiline.printSchema()
```

##IMPORTANT:

primitivesAsString(consider all cols as string), prefersDecimal, allowComments, allowUnquotedFieldNames, allowSingleQuotes, lineSep, samplingRatio, dropFieldIfAllNull, modifiedBefore, modifiedAfter, useUnsafeRow(This is performance optimization when the data is loaded into spark memory) 

```python
df_json=spark.read.json("/Volumes/workspace/default/volumewe47_datalake/simple_json.txt",samplingRatio=0.1,prefersDecimal=True)
display(df_json)
df_json.printSchema()
```
