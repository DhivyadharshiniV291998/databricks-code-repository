#THIS NOTEBOOK-CONTAINS BASIC WRITE OPERATIONS

##Let's get some data we have already...

```python
df1=spark.read.csv("/Volumes/we47catalog/we47schema/we47volume/we47dir1/custs_header",header=True,inferSchema=True)
display(df1.limit(10))
```

##let's write data in different file formats

###modes in writing

- Append - Adds the new data to the existing data. It does not overwrite anything.<br>
- Overwrite - Replaces the existing data entirely at the destination.<br>
- ErrorIfexist(default) - Throws an error if data already exists at the destination.<br>
- Ignore - Skips the write operation if data already exists at the destination.

###1.CSV format

```python
df1.write.csv("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/csv_targetdata",header=True,sep="~",mode="overwrite")
```

###2.json format

```python
df1.write.json("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/json_targetdata",mode="append")
```

###3.orc format

```python
df1.write.orc("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/orc_targetdata",mode="ignore")
spark.read.orc("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/orc_targetdata/").show(2)
```

###4.parquet format

```python
#maxRecordsPerFile is a Spark write option used to control how many records go into each output file when writing data.
df1.write.option("maxRecordsPerFile",10).parquet("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/parquet_targetdata",mode="overwrite",compression="gzip")
```

###5.delta format

```python
df1.write.format("delta").save("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/delta_targetdata",mode="overwrite")
spark.read.format("delta").load("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/delta_targetdata").show(2)
```

###6.default format

```python
df1.write.save("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/what_tragetdata",mode="overwrite")
```

###7.xml format

```python
#rowTag is an option used when reading XML files in Spark to tell Spark which XML element represents one row (one record) in the DataFrame.
df1.write.xml("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/xml_targetdata",mode="overwrite",rowTag="customer")
```

```python
spark.read.xml("/Volumes/we47catalog/we47schema/we47volume/serialized_compressed_data_sources/xml_targetdata",rowTag="customer").show(2)
```

##Table Load Operations - Building LAKEHOUSE ON TOP OF DATALAKE

###1.using saveasTable function

```python
df1.write.saveAsTable("default.customertbl",mode="overwrite")
spark.read.table("default.customertbl").show(10)
```

###2.using insertinto function

```python
#insertInto() is a Spark DataFrameWriter method used to insert data into an existing table.
df1.write.insertInto("default.customertbl",overwrite=True)
spark.read.table("default.customertbl").show(10)
```
